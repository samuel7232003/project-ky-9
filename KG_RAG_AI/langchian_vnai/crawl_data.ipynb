```python
# lấy được dữ liệu pdf từ webpage
# ni giống kiểu crawl data, ổn
import bs4
from langchain_community.document_loaders import WebBaseLoader

web_paths = ["https://rdsic.edu.vn/blog/blog-4/apple-scab-vi-cb.html"]
classes = ['title_center', 'post_content']

# web_paths = ["https://en.wikipedia.org/wiki/Apple_scab"]
# classes = ['mw-content-ltr mw-parser-output']

bs4_strainer = bs4.SoupStrainer(class_=classes)

web_loader = WebBaseLoader(
    web_paths=web_paths,
    bs_kwargs=dict(
        parse_only=bs4_strainer,
    ),
)

docs = web_loader.load()
docs
```

### document splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
chunk_size = 300
chunk_overlap = 30
separators = ['\n\n', '\n', ' ', '']

char_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    length_function=len,
    is_separator_regex=False
)
```

```python
text_out = char_splitter.split_documents(docs)
len(text_out)
```
